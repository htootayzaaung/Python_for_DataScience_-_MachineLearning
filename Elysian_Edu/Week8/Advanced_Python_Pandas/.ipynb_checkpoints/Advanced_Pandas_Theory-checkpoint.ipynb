# Query

## dataframe query

**Dataframe.query()**

```
- expr – expression takes conditions to query rows
- inplace – Defaults to False. When set toTrue, it updates the referring DataFrame and query() method returns None.
- **kwargs –  Keyword arguments that works with eval()
```

````python
## build the dataframe
import pandas as pd
import numpy as np
technologies= {
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas"],
    'Fee' :[22000,25000,23000,24000,26000],
    'Duration':['30days','50days','30days', None,np.nan],
    'Discount':[1000,2300,1000,1200,2500]
          }
df = pd.DataFrame(technologies)
df
````

````python
df2=df.query("Courses == 'Spark'")
df2
````

````python
# Query Rows by using Python variable
value='Spark'
df2=df.query("Courses == @value")
df2
````

````python
# Replace current esisting DataFrame
df.query("Courses == 'Spark'",inplace=True)
df
````

````python
# not equals condition
df2=df.query("Courses != 'Spark'")
df2
````

##  Select Rows Based on List of Column Values

````python
print(df.query("Courses in ('Spark','PySpark')"))
````

````python
# Query Rows by list of values
values=['Spark','PySpark']
print(df.query("Courses in @values"))
````

````python
# Query Rows not in list of values
values=['Spark','PySpark']
df.query("Courses not in @values")
````

## Query with Multiple Conditions

````python
df.query("Fee >= 23000 and Fee <= 24000")
````

# Pivot Table

Using a panda’s pivot table can be a good alternative because it is:

- Quicker
- Easy to use to generate a report or email
- More flexible because it can define custom aggregation functions

````python
import pandas as pd
df = pd.read_excel("resources/sales-funnel.xlsx")
df.head()
````

````python
df.info()
````

````python
df.Status.value_counts()
````

````python
df["Status"] = df["Status"].astype("category")
df["Status"].cat.set_categories(["won","pending","presented","declined"],inplace=True)
````

````python
df.info()
````

````python
df.head(2)
````

````python
pd.pivot_table(df,index=["Name"])
````

## Multiple Index

````python
pd.pivot_table(df,index=["Name","Rep","Manager"])
````

````python
pd.pivot_table(df,index=["Manager","Rep"])
````

## Adding values

````python
## remove the unnecessary columns

pd.pivot_table(df,index=["Manager","Rep"],values=["Price"])
````

## Adding values with aggfunc

````python
# The price column automatically averages the data but we can do a count or a sum. Adding them is simple using aggfunc.
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=np.sum)
````

## Adding multiple values

````python

pd.pivot_table(df,index=["Manager","Rep"],values=["Price"])
````

````python
# aggfunc can take a list of functions
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=[np.mean,len])
````

## Adding expanded values

````python
# If we want to see sales broken down by the products, the columns variable allows us to define one or more columns.
df.head()
````

````python
df.Product.value_counts()
````

````python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=["Product"],aggfunc=[np.sum])
````

````python
# remove the NaN values

pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=["Product"],aggfunc=[np.sum],fill_value=0)
````

````python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price","Quantity"],
               columns=["Product"],aggfunc=[np.sum],fill_value=0)
````

````python
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],aggfunc=[np.sum],fill_value=0)
````

## margin = True

````python
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum,np.mean],fill_value=0)
````

````python
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
````

````python
pd.pivot_table(df,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":np.sum},fill_value=0)
````

````python
table = pd.pivot_table(df,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},fill_value=0)
table
````

## Advanced Pivot Table Filtering


````python
table.query('Manager == ["Debra Henley"]')
````

````python
table.query('Status == ["pending","won"]')
````

# Pandas's melt

The Pandas melt() function is within many other methods used to reshape the pandas DataFrames from wide to a long format which is particularly useful in data science. However, the pd.melt() function is the most efficient and flexible among them. The pd.melt() function unpivots/melts the pandas DataFrame from a wide to a long format.

<img src="resources/melt.png">

## wide vs long dataframe

<img src="resources/wide_df.png">
<img src="resources/long_df.png">

````python
#Raw data in form of dictionary:
data = {"Person":["Alan","Berta","Charlie","Danielle"], #Name of Person
        "House":["A","B","A","C"],                      #Name of houses they live in
        "Age":[32,46,35,28],                            #Age of Person
        "Books":[100,30,20,40],                         #Number of books owned
        "Movies":[10,20,80,60]                          #Number of movie watched
        }

#Converting the raw data into pandas DataFrame:
data_wide = pd.DataFrame(data)
````

````python
data_wide
````

## without any parameters

````python
data_wide.melt()
````

## with id_vars

````python
data_wide.head()
````

````python
data_wide.melt(id_vars=["Person"])
````

## with value_vars

````python
data_wide.melt(id_vars=["Person","House"],           #Identifier columns
               value_vars=["Age","Books","Movies"])
````

````python
data_wide.melt(id_vars=["Person"],             #Identifier columns
               value_vars=["Books","Movies"])
````

````python
data_wide.melt(id_vars=["Person","House"],          #Identifier columns
               value_vars=["Age","Books","Movies"], #Columns to be melted
               var_name="Info",                     #Renaming the variable column name
               value_name="Numerical") 
````

# Dataframe assign()

Dataframe.assign() method assign new columns to a DataFrame, returning a new object (a copy) with the new columns added to the original ones. Existing columns that are re-assigned will be overwritten.

````python
import pandas as pd

df = pd.read_csv("resources/nba.csv")
df.head()
````

````python
# increase the salary by 10 %
df.assign(Revised_Salary = lambda x: df['Salary'] + df['Salary']/10)
````

````python
#  Assigning more than one column at a time
# importing pandas as pd
import pandas as pd

# Making data frame from the csv file
df = pd.read_csv("resources/nba.csv")

# First column ='New_Team', this column
# will append '_GO' at the end of each team name.
# Second column ='Revised_Salary' will increase
# the salary of all employees by 10 %
df.assign(New_team = lambda x: df['Team']+'_GO',Revised_Salary = lambda x: df['Salary'] + df['Salary'] / 10)
````

# Pandas Zip

````python
import pandas as pd
df = pd.DataFrame({
    "fruit":["apple","orange","pine apple"],
    "price 1":[300,400,500],
    "price 2":[100,300,600]
})
````

````python
df["total price"] = list(zip(df["price 1"],df["price 2"]))
````

````python
df
````

````python
def convert_int_to_float(row):
    #idx = row["idx"]
    price1 = float(row["price 1"])
    price2 = float(row["price 2"])
    return price1,price2
````

````python
df["price 1 float"],df["price 2 float"] = zip(*df.apply(convert_int_to_float, axis=1))
````

````python
df
````

````python
import pandas as pd
import numpy as np
  
def color_negative_red(val):
    color = 'blue' if val > 90 else 'black'
    return 'color: % s' % color
  
# creating a DataFrame
dict = {'Maths' : [87, 91, 97, 95],
        'Science' : [83, 99, 84, 76]}
df = pd.DataFrame(dict)
  
# displaying the DataFrame
df.style.applymap(color_negative_red)
````
