# Bubble Chart

A bubble plot is a scatterplot where the circle size is mapped to the value of a third numeric variable. This section shows many bubble plots made with Python, using both the Matplotlib and Seaborn libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt


np.random.seed(42)
N = 100
x = np.random.normal(170, 20, N)
y = x + np.random.normal(5, 25, N)
colors = np.random.rand(N)
area = (25 * np.random.rand(N))**2
```

```python
df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Colors': colors,
    "bubble_size":area})
```

```python
df.head()
```

```python
# scatter plot with scatter() function
plt.scatter('X', 'Y', data=df)
plt.xlabel("X", size=16)
plt.ylabel("y", size=16)
plt.title("Scatter Plot with Matplotlib", size=18)
```

```python
# scatter plot with scatter() function
# transparency with "alpha"
# bubble size with "s"
plt.scatter('X', 'Y', 
             s='bubble_size',
             alpha=0.5, 
             data=df)
plt.xlabel("X", size=16)
plt.ylabel("y", size=16)
plt.title("Bubble Plot with Matplotlib", size=18)
```

```python
# scatter plot with scatter() function
# transparency with "alpha"
# bubble size with "s"
# color the bubbles with "c"
plt.scatter('X', 'Y',
             s='bubble_size',
             c='Colors',
             alpha=0.5, data=df)
plt.xlabel("X", size=16)
plt.ylabel("y", size=16)
plt.title("Bubble Plot with Colors: Matplotlib", size=18)
```
