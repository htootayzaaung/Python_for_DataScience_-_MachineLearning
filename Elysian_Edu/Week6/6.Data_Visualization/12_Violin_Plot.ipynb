# Violin plot
    - Violin plot is a better chart than boxplot as it gives a much broader understanding of the distribution. 
    - It resembles a violin and dense areas point the more distribution of data otherwise hidden by box plots
    - When to use: Its an extension to boxplot. 
    - It should be used when we require a better intuitive understanding of data.
    
 *similar to the boxplot,except they can show the destiny of the data points around a particular value with their widths*

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

exam_data = pd.read_csv('datasets/exams.csv')
```

```python
exam_scores = exam_data[['math score', 'reading score', 'writing score']]
exam_scores.head()
```

```python
# For plotting in boxplot, we convert it to an array
exam_scores_array = np.array(exam_scores)
```

```python
import matplotlib.pyplot as plt
vp = plt.violinplot(exam_scores_array, showmedians = True)
colors = ['blue','grey','lawngreen']
plt.xticks([1,2,3], ['Math','Reading','Writing'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor = colors[i])
    
plt.show()
```
