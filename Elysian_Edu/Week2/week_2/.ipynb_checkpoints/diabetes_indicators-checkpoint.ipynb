# Features selection

## select weighted columns

```python
sel_col = df_corr[df_corr.Diabetes_binary>0.1].index.tolist()
sel_col
```

```python
df_sel = df_input[sel_col]
```

## remove outliers

```python
non_categorical_col = [col for col in df_input.columns if df_input[col].max() > 2]
```

```python
non_categorical_col
```

```python
plt.figure(figsize = (15,15))
for i,col in enumerate(non_categorical_col):
    plt.subplot(4,2,i+1)
    sns.boxplot(x = col, data = df_input)
plt.show()
```

```python
def get_outlier(col_name):
    col_dict = df_input[col_name].value_counts().to_dict()
    outlier = []
    for key, value in col_dict.items():
        if (value/sum(col_dict.values()))*100 < 2:
        #if value < 100:
            outlier.append(key)
    return outlier
```

```python
outlier_index = []
for col in non_categorical_col:
    outlier = get_outlier(col)
    index_list = list(df_input[df_input[col].isin(outlier)].index)
    outlier_index.append(index_list)
outlier_index = [i for index in outlier_index for i in index]
outlier_index = list(set(outlier_index))
```

```python
df_no_outlier = df_input[~df_input.index.isin(outlier_index)]
```

```python
df_no_outlier.hist(figsize=(30,20))
```

## all features

```python
#baseline
X = df_input.drop(columns = "Diabetes_binary")
y = df_input.Diabetes_binary
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.01, random_state=42, stratify=y)
```

```python
X_train.reset_index(drop=True,inplace=True)
y_train.reset_index(drop=True,inplace=True)
```

```python
scale_col = [col for col in X_train.columns if X_train[col].max() > 1]
```

```python
r_scaler = RobustScaler()
```

```python
r_scaler = r_scaler.fit(X_train[scale_col])
X_train[scale_col] = r_scaler.transform(X_train[scale_col])
```

## selected features

```python
X_sel = df_sel.drop(columns = "Diabetes_binary")
y_sel = df_sel.Diabetes_binary
```

```python
X_train_sel, X_test_sel, y_train_sel, y_test_sel = train_test_split(X_sel,y_sel, test_size=0.01, random_state=42, stratify=y_sel)
```

```python
X_train_sel.reset_index(drop=True,inplace=True)
y_train_sel.reset_index(drop=True,inplace=True)
```

```python
scale_col_sel = ['BMI', 'GenHlth',  'PhysHlth', 'Age']
```

```python
r_scaler_sel = RobustScaler()
```

```python
r_scaler_sel = r_scaler_sel.fit(X_train_sel[scale_col_sel])
X_train_sel[scale_col_sel] = r_scaler_sel.transform(X_train_sel[scale_col_sel])
```

# Sampling

```python
from imblearn.over_sampling import SMOTENC
smote_nc = SMOTENC(categorical_features=[0, 1],random_state=42)
```

```python
X_sm, y_sm = smote_nc.fit_resample(X_train, y_train)
```

```python
smote_nc_sel = SMOTENC(categorical_features=[0, 1],random_state=42)
```

```python
X_sm_sel, y_sm_sel = smote_nc_sel.fit_resample(X_train_sel, y_train_sel)
```

```python
print(f'''Shape of X before SMOTE: {X_train.shape}
Shape of X after SMOTE: {X_sm.shape}''')

print('\nBalance of positive and negative classes (%):')
y_sm.value_counts(normalize=True) * 100
```

# Model training

## logistic regression

```python
model_log = LogisticRegression()
model_log_sel = LogisticRegression()
```

```python
model_log.fit(X_sm, y_sm)
```

```python
model_log_sel.fit(X_sm_sel, y_sm_sel)
```

##  XGBoost

```python

model_xgb = XGBClassifier()
model_xgb_sel = XGBClassifier()
```

```python
model_xgb.fit(X_sm, y_sm)
```

```python
model_xgb_sel.fit(X_sm_sel, y_sm_sel)
```

# prediction and evaluation

```python
X_test[scale_col] = r_scaler.transform(X_test[scale_col])
```

```python
X_test_sel[scale_col_sel] = r_scaler_sel.transform(X_test_sel[scale_col_sel])
```

## logistic regression

```python
pred_log = model_log.predict(X_test)
#pred_log_sel = model_log_sel.predict(X_test_sel)
```

```python
pred_log_sel = model_log_sel.predict(X_test_sel)
```

```python
print(classification_report(y_test, pred_log))
print('----------------------------------------------')
print(classification_report(y_test_sel, pred_log_sel))
```

```python
print(classification_report(y_test, pred_log))
print('----------------------------------------------')
print(classification_report(y_test_sel, pred_log_sel))
```

According to the classification report, the diffenence in accuracy between the two models ( with 22 features and with 15 features ) is only 0.01, which means the later model have the better perfomance as it needs less features than the previous.

## XGBoost

```python
pred_xgb = model_xgb.predict(X_test)
pred_xgb_sel = model_xgb_sel.predict(X_test_sel)
```

```python
print(classification_report(y_test, pred_xgb))
print('----------------------------------------------')
print(classification_report(y_test, pred_xgb_sel))
```

```python
print(classification_report(y_test, pred_xgb))
print('----------------------------------------------')
print(classification_report(y_test, pred_xgb_sel))
```

```python
cm = confusion_matrix(y_test, pred_xgb_sel)    

ax= plt.subplot()
sns.heatmap(cm, cmap="YlGnBu",annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['no diabetes', 'diabetes']); ax.yaxis.set_ticklabels(['no diabetes', 'diabetes']);
```
